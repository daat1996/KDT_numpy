```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%precision 3
```

```python
from scipy import integrate
import warnings
```

```python
x_range = np.array([0,1])
```

```python
def f(x):
    if x_range[0] <= x <=x_range[1]:
        return 2*x
    else:
        return 0
```

```python
X = [x_range, f]
```

불공정한 룰렛에 대한 밀도함수와 확률

```python
xs = np.linspace(x_range[0], x_range[1], 100)

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.plot(xs,[f(x) for x in xs], label='f(x)', color='gray')
ax.hlines(0, -0.2, 1.2, alpha=0.3)
ax.vlines(0, -0.2, 2.2, alpha=0.3)
ax.vlines(xs.max(),0,2.2, linestyles=':', color='gray')

# 0.4 부터 0.6까지 x좌표를 준비
xs= np.linspace(0.4, 0.6, 100)
# xs의 범위로 f(x)와 x축으로 둘러싸인 영역에 색을 적용
ax.fill_between(xs, [f(x) for x in xs], label='prob11')

ax.set_xticks(np.arange(-0.2, 1.3, 0.1))
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.2, 2.1)
ax.legend()

plt.show()
```

이 적분은 integrate의 quad 함수를 사용하여 구할수 있다.

```python
# 첫 번째 인수는 피적분함수, 두 번째 인수와 세 번째 인수는 적분 범위
integrate.quad(f, 0.4, 0.6)
```

minimize_scalar는 함수의 최솟값을 구하는 함수로 이것을 사용하여 f(x)의 최솟값이 0이상으로 나오면 f(x)가 항상 0 이상의 값을 취한다는 의미를 갖는다

```python
from scipy.optimize import minimize_scalar
res = minimize_scalar(f)
# 함수의 최솟값은 fun이라는 인스턴스 변수에
res.fun
```

numpy에서는 무한대를 np.inf로 나타낼 수 있다.

```python
integrate.quad(f, -np.inf, np.inf)[0]
```
